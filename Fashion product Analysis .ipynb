```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Exploratory Data Analysis  

```python
ax=sns.countplot(x="Product Name",data=df,)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Product Name'],as_index=False)['Price'].sum().sort_values('Price',ascending=False )
sns.barplot(x='Product Name', y= 'Price',data=sales_gen)
```

## From above graph most selling product are Jeans. 

```python
sales_brand=df.groupby(['Brand'],as_index=False)['Price'].sum().sort_values('Price',ascending=False )
sns.barplot(x='Brand', y= 'Price',data=sales_brand)
```

## Top 3 selling brand are Nike, H&M and Zara respectively

```python
ax=sns.countplot(x="Rating",data=df,)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_brand=df.groupby(['Brand'],as_index=False)['Rating'].sum().sort_values('Rating',ascending=False )
sns.barplot(x='Brand', y= 'Rating',data=sales_brand)
```

# Most likable products by customers are Nike

```python
sales_brand=df.groupby(['Category','Product Name'],as_index=False)['Price'].sum().sort_values('Price',ascending=False )
sns.barplot(x='Category', y= 'Price',data=sales_brand ,errorbar=None)
```

```python
sales_brand=df.groupby(['Category','Brand'],as_index=False)['Price'].sum().sort_values('Price',ascending=False )
sns.barplot(x='Brand', y= 'Price',data=sales_brand,errorbar=None)
```

# From above most selling category kids fashion category by Nike
